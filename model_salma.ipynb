```python
# --- ISI DI DALAM FILE training_model.ipynb ---
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
import pickle

# 1. Membuat data dummy untuk 4 fitur input
np.random.seed(42)
X = np.random.rand(100, 4)
y = np.random.randint(0, 2, 100)

# 2. Membuat dan mengekspor scaler.pkl
scaler = StandardScaler()
scaler.fit(X)
with open('scaler.pkl', 'wb') as f:
    pickle.dump(scaler, f)

# 3. Membuat dan mengekspor model.pkl
X_scaled = scaler.transform(X)
model = LogisticRegression()
model.fit(X_scaled, y)
with open('model.pkl', 'wb') as f:
    pickle.dump(model, f)
    
print("Sukses! File model.pkl dan scaler.pkl telah berhasil dibuat.")
```
